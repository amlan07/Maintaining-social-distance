```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
import threading
import winsound
freq=1500
dur=500
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.namedWindow ('screen', cv2.WINDOW_NORMAL)
cv2.setWindowProperty ('screen', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

def face_lines(img,x1,y1,x2,y2):
    diagonal = math.sqrt(math.pow((x1-x2),2)+math.pow((y1-y2),2))
    p=int(0.1712328*diagonal)
    cv2.line(img,(x1,y1),(x1,y1+p),(255,255,255),1)
    cv2.line(img,(x1,y2-p),(x1,y2),(255,255,255),1) 
    
    cv2.line(img,(x2,y1),(x2,y1+p),(255,255,255),1)    
    cv2.line(img,(x2,y2-p),(x2,y2),(255,255,255),1)
    
    cv2.line(img,(x1,y1),(x1+p,y1),(255,255,255),1)
    cv2.line(img,(x2-p,y1),(x2,y1),(255,255,255),1)    
    cv2.line(img,(x1,y2),(x1+p,y2),(255,255,255),1)
    cv2.line(img,(x2-p,y2),(x2,y2),(255,255,255),1)
    return img


class beeper(threading.Thread):
    def run(self):
        winsound.Beep(freq, dur)
            


car_cascade=cv2.CascadeClassifier('haarcascade_fullbody.xml')
#car2_cascade=cv2.CascadeClassifier('haarcascade_upperbody.xml')
cap = cv2.VideoCapture('vtest.avi')
while(cap.isOpened()):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.putText(frame,"Maintain Social distance due to Covid 19",(40,50),font, 1.1,(0,0,255),2)
    cars=car_cascade.detectMultiScale(gray,1.05,2)
    cars_new=[]
    for (x,y,w,h) in cars:
        gray2=gray[y:y+h,x:x+w]
        cars2=car_cascade.detectMultiScale(gray2,1.05,2)
        if len(cars2)==0:
            continue
        cars_new.append((x,y,w,h))
    for j in range (len(cars_new)):
        i=cars_new[j]
        x=i[0]
        y=i[1]
        w=i[2]
        h=i[3]
        frame=face_lines(frame,x,y,x+w,y+h)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        for k in range (j+1,len(cars_new)):
            i1=cars_new[k]
            x1=i1[0]
            y1=i1[1]
            w1=i1[2]
            h1=i1[3]
            d=math.sqrt(abs((x+w/2)-(x1+w1/2))**2+abs((y+h)-(y1+h1))**2)
            d=round(d)
            if d<55:
                beep=beeper()
                beep.start()
                cv2.putText(frame,"Take precautions",(40,500),font, 0.8,(0,0,255),2)
                cv2.line(frame,(x+int(w/2),y+h),(x1+int(w1/2),y1+h1),(0,0,255),3)
                cv2.putText(frame,str(d)+'m',(int(((x+w/2)+(x1+w1/2))/2),int(((y+h)+(y1+h1))/2)),font, 0.4,(0,0,255),2)
            else:
                pass
                #cv2.line(frame,(x+int(w/2),y+h),(x1+int(w1/2),y1+h1),(255,255,255),1)
                #cv2.putText(frame,str(d)+'m',(int(((x+w/2)+(x1+w1/2))/2),int(((y+h)+(y1+h1))/2)),font, 0.4,(255,255,255),2)
    cv2.imshow('screen',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
